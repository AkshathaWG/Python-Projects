```python
# Associate utility functions with the numbering of alternatives
V = {1: V1, 2: V2, 3: V3}
prob_e = V1
prob_h = V2
prob_p = V3
```

```python
# Calculation of the cross elasticities.
# We use the 'Derive' operator to calculate the derivatives.
cross_elas_elec_kms = Derive(prob_e, 'kms') * kms / prob_e
cross_elas_hyb_kms = Derive(prob_h, 'kms') * kms / prob_h
cross_elas_pet_kms = Derive(prob_p, 'kms') * kms / prob_p

cross_elas_elec_hsize = Derive(prob_e, 'hsize') * hsize / prob_e
cross_elas_hyb_hsize = Derive(prob_h, 'hsize') * hsize / prob_h
cross_elas_pet_hsize = Derive(prob_p, 'hsize') * hsize / prob_p
```

```python
# simulatedValues is a Panda dataframe with the same number of rows as
# the database, and as many columns as formulas to simulate.
simulatedValues = biogeme.simulate(results.getBetaValues())

# We calculate the elasticities
simulatedValues['Weighted prob elec'] = (
    simulatedValues['weight'] * simulatedValues['prob_e']
)

denominator_e = simulatedValues['Weighted prob elec'].sum()

cross_elas_term_e = (
    simulatedValues['Weighted prob elec']
    * simulatedValues['cross_elas_elec_kms']
    / denominator_e
).sum()

print(
    f'Aggregate cross elasticity of Electric Car wrt VKTdaily: '
    f'{cross_elas_term_e:.3g}'
)
```
